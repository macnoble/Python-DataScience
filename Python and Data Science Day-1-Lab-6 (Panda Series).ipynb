```python
import numpy as np
import csv
import datetime
import pandas as pd 
```

```python
def openCSV(filename):
    with open(filename, newline='') as file:
        reader = csv.DictReader(file)
        result = np.empty((0,2))
        for row in reader:
            headerRow = np.array([[row['Date'], row['Volume']]])
            result = np.append(result, headerRow, axis=0)
    return result

```

```python
appleStockPrices2016 = openCSV('AAPL-2016.csv')
print(appleStockPrices2016)
```

```python
date = [datetime.datetime.strptime(x,'%Y-%m-%d') for x in appleStockPrices2016[:,0]]
data = pd.Series(np.float64(appleStockPrices2016[:,1]), index=date)
data.plot()
```
